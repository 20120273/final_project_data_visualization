## Tiền xử lý dữ liệu để đưa về dạng có thể phân tích

DÒng đầu tiên là câu hỏi, ta sẽ lấy dòng này lưu vào `dic_question` và xoá dòng này đi

```python
dic_question = dict(df_1.iloc[0,:])
df_1 = df_1.drop(0)
```

Chuyển `dic_question` thành `new_dic` bằng việc bỏ những câu hỏi trùng nhau

```python
new_dic = dict()
for i in dic_question.keys():
    i1 = i + ' '
    
    if '_1 ' in i1:
        if "?" in dic_question[i]:
            new_dic[i[0:i.index("_")]] = dic_question[i][0:dic_question[i].index('?')+1]
        elif ":" in dic_question[i]: 
            new_dic[i[0:i.index("_")]] = dic_question[i][0:dic_question[i].index(':')]
        else:
            new_dic[i[0:i.index("_")]] = dic_question[i]
    elif '_' not in i:
        if "?" in dic_question[i]:
            new_dic[i] = dic_question[i][0:dic_question[i].index('?')+1]
        elif ":" in dic_question[i]: 
            new_dic[i] = dic_question[i][0:dic_question[i].index(':')]
        else:
            new_dic[i] = dic_question[i]
new_dic['Duration (in seconds)'] = 'Duration'
new_dic
```

Vì mục đích của nhóm là phân tích ở 4 quốc giá là **Japan, China, South Korea, Viet Nam** nên sẽ chỉ lấy dưng dòng thuộc 4 quốc gia này.

```python
df_1 = df_1[(df_1['Q4'] == "Japan") | (df_1['Q4'] == "China") | (df_1['Q4'] == "South Korea") | (df_1['Q4'] == "Viet Nam")]
df_1.head(5)
```

Columnss có 2 dạng, loại 1 là câu hỏi trả lời đáp án, loại 2 là những câu hỏi _1, _2, ... là dạng catogerial nên sẽ chọn những câu loại 1 lưu trước vào `df`.

```python
col = df_1.columns
newcol = []
for i in col:
    if '_' not in i:
        newcol.append(i)
```

```python
df = df_1[newcol]
df.head(5)
```

Tiếp đến, ta sẽ xử lý những câu hỏi loại 2 và tiếp tục lưu nó vào `df`

```python
for i in range(len(col)):
    ij = col[i] + ' ' 
    if '_1 ' in ij:
        arr = [elem for elem in col if elem.startswith(col[i][0:col[i].index("_")])]
        temp_df = df_1[arr]
        a = [tuple(filter(lambda x: pd.notnull(x), row)) for row in temp_df.values]
        df[col[i][0:col[i].index("_")]] = a
df = df.applymap(lambda x: np.nan if isinstance(x, tuple) and len(x) == 0 else x)
```

Vì name của columns đang ở dạng Q1, Q2, ... nên ta sẽ chuyển thành câu hỏi và reindex để dữ liệu dễ phân tích hơn.

```python
sort_col = ['Duration (in seconds)']
for i in range(2,45):
    sort_col.append('Q' + str(i))
df = df.reindex(columns=sort_col)
df = df.reset_index(drop=True)
df.rename(columns=new_dic, inplace = True)
df.index = df.index + 1
df
```

## XỬ LÝ DỮ LIỆU NAN

Ta xem thử mỗi cột có bao nhiêu dữ liệu nan

```python
df.isnull().sum()
```

- Với 6 câu hỏi (columns) đầu tiên, ta không thấy xuất hiện giá trị NAN. Có lẽ đây là những câu trả hỏi bắc buộc của khảo sát nên rất đầy đủ giá trị.

- Với các câu hỏi (columns) còn lại, ta thấy xuất hiện rất nhiều giá trị NAN. Vì đây là những câu hỏi tự chọn. Ta sẽ xoá những cột có quá nhiều giá trị NAN vì những cột này sẽ mang lại rất ít thông tin những lại gây khó khăn nhiều trong quá trình phân tích. Với dữ liệu hơn 1500 dòng, ta sẽ xoá những dòng có hơn 500 dòng NAN.

```python
nan_counts = df.isna().sum()

columns_to_drop = nan_counts[nan_counts > 500].index

df.drop(columns=columns_to_drop, inplace= True)
df.isna().sum()
```

Giờ dữ liệu chỉ còn những cột có ít hơn 500 dòng chứa NAN.
